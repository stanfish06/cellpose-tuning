```python
import torch
from torchvision.transforms import v2
import matplotlib.pyplot as plt
import numpy as np
import albumentations as A
from albumentations.pytorch import ToTensorV2
import cv2
import os
import pickle 
import cellpose
from cellpose import core, utils, io, models, metrics
import shutil
import tifffile as tf
```

```python
img = tf.imread('./data/micropatterned_colony/train_window/z8.tif')
mask = np.load('./data/micropatterned_colony/train_window/z8_seg.npy', allow_pickle=True).item()['masks']
plt.imshow(img[..., 1])
```

```python
sub_img_width = 100
sub_img_height = 100
transforms = A.Compose([
    A.ShiftScaleRotate(
        shift_limit = 0, 
        scale_limit = 0, 
        rotate_limit = 360, 
        p = 1, 
        interpolation=cv2.INTER_NEAREST
    ),
    A.RandomCrop(sub_img_width, sub_img_height),
    ToTensorV2(),
])
targets = {'mask': 'image'}
co_transforms = A.Compose(transforms, p = 1, additional_targets = targets)
```

```python
augmented_data_path = './data/micropatterned_colony/train_window/train_data'
if os.path.exists(augmented_data_path):
    shutil.rmtree(augmented_data_path)  # Delete the folder if it exists

os.makedirs(augmented_data_path)  # Recreate the folder
next_ind = 0
```

```python
augment_size = 200
min_nonzero_prop = 0.5
for i in range(augment_size):
    new_imgs = co_transforms(image = img.astype(np.float64), mask = mask.astype(np.float64))
    sub_img = new_imgs['image']
    sub_mask = new_imgs['mask'][0, ...]
    nonzero_prop = np.count_nonzero(sub_mask) / (sub_img_width * sub_img_height)
    if nonzero_prop > min_nonzero_prop:
        tf.imwrite(os.path.join(augmented_data_path, f'_{next_ind}.tif'), np.stack(np.vstack([sub_img.numpy(), np.zeros((1, 100, 100))]), -1).astype('uint16'))
        np.save(os.path.join(augmented_data_path, f'_{next_ind}_seg.npy'), np.array({'masks': sub_mask.numpy().astype('uint16')}))
        next_ind = next_ind + 1
```

```python
augmented_data_path = './data/micropatterned_colony/train_window/test_data'
if os.path.exists(augmented_data_path):
    shutil.rmtree(augmented_data_path)  # Delete the folder if it exists

os.makedirs(augmented_data_path)  # Recreate the folder
next_ind = 0
```

```python
test_size = 50
min_nonzero_prop = 0.5
for i in range(test_size):
    new_imgs = co_transforms(image = img.astype(np.float64), mask = mask.astype(np.float64))
    sub_img = new_imgs['image']
    sub_mask = new_imgs['mask'][0, ...]
    nonzero_prop = np.count_nonzero(sub_mask) / (sub_img_width * sub_img_height)
    if nonzero_prop > min_nonzero_prop:
        tf.imwrite(os.path.join(augmented_data_path, f'_{next_ind}.tif'), np.stack(np.vstack([sub_img.numpy(), np.zeros((1, 100, 100))]), -1).astype('uint16'))
        np.save(os.path.join(augmented_data_path, f'_{i}_seg.npy'), np.array({'masks': sub_mask.numpy().astype('uint16')}))
        next_ind = next_ind + 1
```
